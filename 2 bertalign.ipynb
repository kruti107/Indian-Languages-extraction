```python
import os
import fitz

def convert_pdfs_to_images(input_folder, output_folder):
    # Ensure the output folder exists
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Iterate through each folder and subfolder in the input directory
    for root, dirs, files in os.walk(input_folder):
        # Create corresponding subfolders in the output directory
        relative_path = os.path.relpath(root, input_folder)
        output_subfolder = os.path.join(output_folder, relative_path)
        os.makedirs(output_subfolder, exist_ok=True)

        # Iterate through each PDF file in the current folder
        for file_name in files:
            if file_name.endswith('.pdf'):
                # Path to the input PDF file
                pdf_path = os.path.join(root, file_name)

                # Open the PDF file
                pdf_document = fitz.open(pdf_path)

                # Convert each page to PNG
                for page_number in range(len(pdf_document)):
                    page = pdf_document.load_page(page_number)
                    image = page.get_pixmap()
                    image_path = os.path.join(output_subfolder, f'{os.path.splitext(file_name)[0]}_page_{page_number+1}.png')
                    image.save(image_path)
                    print(f'Page {page_number+1} of {file_name} saved as {image_path}')

                # Close the PDF document
                pdf_document.close()

# Example usage:
input_folder = 'D://desktop//exploreCSR//codeos//hindi//text1'
output_folder = 'D://desktop//exploreCSR//codeos//hindi//images_text1'

convert_pdfs_to_images(input_folder, output_folder)
```

```python
import os
import pytesseract
from PIL import Image

# Set the path to the Tesseract executable
pytesseract.pytesseract.tesseract_cmd = "D://desktop//tesseract.exe"

def get_text(img_path):
    # Open the image
    image = Image.open(img_path)
    
    # Perform OCR with Hindi language
    text = pytesseract.image_to_string(image, lang='hin')
    
    return text

def process_images_in_folder(input_folder, output_folder):
    # Iterate through files and subdirectories in the given input directory
    for root, dirs, files in os.walk(input_folder):
        # Create corresponding subfolders in the output directory
        relative_path = os.path.relpath(root, input_folder)
        output_subfolder = os.path.join(output_folder, relative_path)
        os.makedirs(output_subfolder, exist_ok=True)

        # Iterate through each image file in the current folder
        for file_name in files:
            if file_name.endswith('.png'):
                # Path to the input image file
                img_path = os.path.join(root, file_name)

                # Get OCR text from the image
                text = get_text(img_path)

                # Path to the output text file
                output_text_path = os.path.join(output_subfolder, f'{os.path.splitext(file_name)[0]}.txt')

                # Write the OCR text to the output text file
                with open(output_text_path, 'w', encoding='utf-8') as text_file:
                    text_file.write(text)
                    print(f'OCR output saved as {output_text_path}')

# Specify the parent directory containing all subfolders with images
parent_input_folder = 'D://desktop//exploreCSR//codeos//hindi//images_text1//'  # Replace with the path to your parent input image folder
output_folder = 'D://desktop//exploreCSR//codeos//hindi//images to text file//'  # Replace with the path to your output text folder

# Process images in the parent input folder
process_images_in_folder(parent_input_folder, output_folder)
```

```python
import os

def merge_text_files(input_folder, output_folder):
    # Iterate through subfolders in the input folder
    for root, dirs, files in os.walk(input_folder):
        # Create corresponding subfolders in the output directory
        relative_path = os.path.relpath(root, input_folder)
        output_subfolder = os.path.join(output_folder, relative_path)
        os.makedirs(output_subfolder, exist_ok=True)

        # Group text files by lecture number
        lecture_files = {}
        for file in files:
            if file.endswith('.txt'):
                parts = file.split('_')
                lecture_num = parts[1]
                if lecture_num not in lecture_files:
                    lecture_files[lecture_num] = []
                lecture_files[lecture_num].append(os.path.join(root, file))

        # Merge text files into a single file for each lecture
        for lecture_num, files in lecture_files.items():
            merged_text = ''
            for file_path in files:
                with open(file_path, 'r', encoding='utf-8') as file:
                    merged_text += file.read() + '\n'  # Add newline between pages
            output_file_path = os.path.join(output_subfolder, f'merged_lecture_{lecture_num}.txt')
            with open(output_file_path, 'w', encoding='utf-8') as output_file:
                output_file.write(merged_text.strip())  # Remove trailing newline
            print(f'Merged text files for Lecture {lecture_num} saved as {output_file_path}')

# Specify the parent directory containing subfolders with text files
parent_input_folder = 'D://desktop//exploreCSR//codeos//hindi//images to text file//'  # Replace with the path to your parent input folder
output_folder = 'D://desktop//exploreCSR//codeos//hindi//merged_tex files//'  # Replace with the path to your output folder

# Merge text files in subfolders of the parent input folder
merge_text_files(parent_input_folder, output_folder)
```
