```python
import os
from pikepdf import Pdf

# Define a function to remove images from a page
def remove_image(page):
    # Check if the page has any images
    if page.images:
        # Get the image names on the page
        image_names = list(page.images.keys())
        # Iterate through each image and remove it
        for image_name in image_names:
            # Remove the image from the page resources
            del page.Resources.XObject[image_name]

# Define a function to process PDFs in a directory
def process_pdfs(directory):
    # Iterate through files and subdirectories in the given directory
    for root, dirs, files in os.walk(directory):
        for file in files:
            # Check if the file is a PDF
            if file.endswith('.pdf'):
                # Construct the full path to the PDF file
                pdf_file = os.path.join(root, file)
                print(f"Processing {pdf_file}")
                # Open the PDF file
                pdf = Pdf.open(pdf_file)
                # Iterate through each page and remove images
                for page in pdf.pages:
                    remove_image(page)
                # Construct the path for the modified PDF file
                modified_pdf_path = os.path.join(root, f"modified_{file}")  # Save with a modified filename
                # Save the modified PDF
                pdf.save(modified_pdf_path)
                print(f"Images removed from {pdf_file}. Modified PDF saved to {modified_pdf_path}")

# Specify the parent directory containing all subfolders with PDF files
parent_directory = 'D://desktop//exploreCSR//codeos//hindi//Text'

# Process PDFs in the parent directory
process_pdfs(parent_directory)
```
